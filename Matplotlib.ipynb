# Data Visualisation

Is the presentation of data in a graphical or pictorial format.It involves creating visual representation of data identify patterns,trends, and insides that may not apparent in raw data.

# Matplotlib

Matplotlib is a comprehensive library for creating static,animated, and interactive visualization in Python.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

A "figure" means the entire image containing all the subplots."Axes" refers to the individual subplots.

```python
a = plt.subplots()
print(a)
```

```python
# You can change figure size
fig,ax = plt.subplots(figsize = (2,3))
```

```python
# lets create a bar graph
fig,ax = plt.subplots(figsize=(5,5))
x = ["Shampoo","Oil","Soap"]
y = [1234,200,800]
ax.bar(x,y)
```

```python
# let's add something more to it
fig,ax = plt.subplots(figsize=(5,5))
x = ["Shampoo","Oil","Soap"]
y = [1234,200,800]
ax.bar(x,y)
ax.set_xlabel("<------ Product ------>")
ax.set_ylabel("<---- Units sold ---->")
ax.set_title("Units Sold For Each Product")
ax.grid(True) # shows grid
```

```python
# let's change the color also
# lets create a bar graph
fig,ax = plt.subplots(figsize=(5,5))
x = ["Shampoo","Oil","Soap"]
y = [1234,200,800]
color_bar = ["red","purple","yellow"]
ax.bar(x,y,color=color_bar)
ax.grid(True)
```

```python
help(ax.bar)
```

```python

# scatter plot
marketing_spend = np.linspace(1,20)
sales = [i**2 for i in marketing_spend]
sales1 = [i**3 for i in marketing_spend]

fig,ax = plt.subplots(figsize=(5,5))
ax.scatter(marketing_spend,sales,alpha=0.8,color="red")
ax.scatter(marketing_spend,sales1,alpha=0.8,color="blue")
ax.set_xlabel("Marketing Spend")
ax.set_ylabel("Sales")
ax.legend(["**2","**3"])
ax.grid(True)
```

```python
# line graph
marketing_spend = np.arange(1,20)
sales = [i**2 for i in marketing_spend]
sales1 = [i**3 for i in marketing_spend]

fig,ax = plt.subplots(figsize=(5,5))
ax.plot(marketing_spend,sales,alpha=0.8,color="red")
ax.plot(marketing_spend,sales1,alpha=0.8,color="blue")
ax.set_xlabel("Marketing Spend")
ax.set_ylabel("Sales")
ax.legend(["**2","**3"])
ax.grid(True)
```

```python
help(ax.plot)
```

```python
# pie chart
labels = ["Samsung","Redmi","OnePlus","OppO"]
size = [30,20,15,35]
fig,ax = plt.subplots(figsize=(5,5))
ax.pie(size,labels=labels,autopct="%1.2f")
ax.set_title("Market Share")
```

```python
help(ax.pie)
```
