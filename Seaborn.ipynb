# Seaborn

Seaborn is library  for making statistical graphics in python. It builds on top of matplotlib and integrates closely  with pandas data structures.

```python
import seaborn as sns
```

```python
help(sns.load_dataset)
```

```python
df = sns.load_dataset("tips")    # try uisng flights dataset
df
```

```python
(sns.set_theme)
```

```python
# bar graph
sns.countplot(data=df,x="sex")
```

```python
sns.set_theme(style="whitegrid")
```

```python
sns.barplot(data=df,x="sex",y="tip")
```

```python
sns.barplot(data=df,x="time",y="total_bill",hue="sex")
```

```python
# Histogram
sns.histplot(data=df,x="total_bill")
```

```python
# scatter plot
sns.scatterplot(data=df,x="total_bill",y="tip",hue="smoker")
```

```python
# line plot
sns.set_theme()
sns.lineplot(data=df,x="total_bill",y="tip")
```

```python
sns.lineplot(data=df,x="total_bill",y="tip",hue="smoker")
```
